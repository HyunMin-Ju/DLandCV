```python
import numpy as np
import time

def assignment1_1_function(SIZE):
  HALF = int(SIZE/2)
  FULL = int(SIZE-1)
  # . . .
  L1 = range(SIZE) # Python list
  L2 = range(SIZE) # Python list

  A1 = np.arange(SIZE) # NumPy array
  A2 = np.arange(SIZE) # NumPy array

  # Time element-wise addition using Python list
  start = time.time()
  list_addition = [x + y for x, y in zip(L1, L2)]# computation code between time.time()s
  end=time.time()
  list_add_time = end - start

  # Time element-wise addition using NumPy array
  start = time.time()
  numpy_addition = A1 + A2# computation code between time.time()s
  end=time.time()
  numpy_addition_time = end - start

  # Time dot product using Python list
  start = time.time()
  list_dot = sum([x * y for x, y in zip(L1, L2)])
  end=time.time() # computation code between time.time()s
  list_dot_time = end - start

  # Time dot product using NumPy array
  start = time.time()
  numpy_dot = np.dot(A1,A2)
  end=time.time()
  numpy_dot_time = end - start

  # printing code
  print("--------------------------------------------------------------------------")
  print("SIZE =",SIZE)
  print("--------------------------------------------------------------------------")
  print("List Addition values at 0, SIZE-1 = ",   list_addition[0]," , ",list_addition[SIZE-1])

  print("List addition time [ms] =", list_add_time*1000)
  print("--------------------------------------------------------------------------")
  print("List dot product = ",list_dot)
  print("List dot product time [ms] = ", list_dot_time*1000)
  print("--------------------------------------------------------------------------")
  print("NumPy Addition values at 0, SIZE-1 = ",   numpy_addition[0]," , ",numpy_addition[SIZE-1])
  print("NumPy addition time [ms] =", numpy_addition_time*1000)
  print("--------------------------------------------------------------------------")
  print("NumPy dot product = ", numpy_dot)
  print("NumPy dot product time [ms] = ", numpy_dot_time*1000)
  print("--------------------------------------------------------------------------")
```

```python
assignment1_1_function(100)
assignment1_1_function(10000)
assignment1_1_function(1000000)
```
